```python
# Carico il dataset in un DataFrame
import pandas as pd

covid_data = pd.read_csv("owid-covid-data.csv")
```

```python
# Verifica dimensioni del datset 
print("Dimensioni del dataset:")
print(covid_data.shape)
```

```python
# Visualizzazione delle prime 5 righe del dataset per capire la struttura
print("Prime cinque righe del dataset:")
covid_data.head()
```

```python
# Verifica colonne dataset
print("Colonne del dataset:")
covid_data.columns
```

```python
# Verifica dei metadati del dataset
print("Metadati del dataset:")
covid_data.info()
```

```python
# Controllo descrittori statistici di base
print("Descrittori statistici di base:")
covid_data.describe()
```

```python
# Iteriamo ogni colonna del DataFrame
for column in covid_data.columns:
    # Verifichiamo se tipo categorico
    if covid_data[column].dtype=="object":
        print(f"{column} column contains categorical data.")
        # Conteggio valori unici
        unique_count = covid_data[column].nunique()
        print(f"{column} contains {unique_count} unique values")
        # Frequenza di ogni valore unico
        all_count = covid_data[column].value_counts()
        print(f"Distribution of every unique value is as such:")
        print(all_count)
        print()
```

```python
# Iteriamo ogni colonna del DataFrame
for column in covid_data.columns:
    # Somma valori nulli
    nan_count = covid_data[column].isna().sum()
    # Calcolo della percentuale
    nan_percentage = round((nan_count/covid_data.shape[0])*100, 2)
    print(f"{column} contains {nan_count} NaN values, {nan_percentage}% of all rows.")
```

```python
#Elimino dati nulli dalla colonna continent
continent_df = covid_data.dropna(subset=['continent'])
    
#Calcolo il valore max per ogni location, essendo il total_cases un valore comulativo
max_cases_per_continent = continent_df.groupby(['continent', 'location'])['total_cases'].max()
    
#Somma dei max di ogni location per ottenere il totale
total_cases = max_cases_per_continent.groupby(level='continent').sum()
    
total_cases.to_frame()
```

```python
import humanize

# Calcolo dei descrittori statistici per i due continenti
df_europe = continent_df[continent_df['continent'] == 'Europe']
df_asia = continent_df[continent_df['continent'] == 'Asia']

# Valore massimo per Europa
max_europe = df_europe.groupby('location')['total_cases'].max().sum()
formatted_max_europe = humanize.intword(max_europe)
print("Valore massimo per Europa:")
formatted_max_europe
```

```python
# Valore massimo per Asia
max_asia = df_asia.groupby('location')['total_cases'].max().sum()
formatted_max_asia = humanize.intword(max_asia)
print("Valore massimo per Asia:")
formatted_max_asia
```

```python
# Media per Europa
mean_europe = df_europe['total_cases'].mean()
formatted_mean_europe = humanize.intword(mean_europe)
print("Media per Europa:")
formatted_mean_europe
```

```python
# Media per Asia
mean_asia = df_asia['total_cases'].mean()
formatted_mean_asia = humanize.intword(mean_asia)
print("Media per Asia:")
formatted_mean_asia
```

```python
# Calcolo numero totale di casi nel mondo
total_world_cases = covid_data['new_cases'].sum()
formatted_total_world_cases = humanize.intword(total_world_cases)
print("Numero totale di casi nel mondo:")
formatted_total_world_cases
```

```python
# Percentuale rispetto al numero totale di casi nel mondo Europa
percentage_europe = (df_europe['new_cases'].sum() / total_world_cases) * 100
print("Percentuale rispetto al numero totale di casi nel mondo Europa:")
percentage_europe
```

```python
# Percentuale rispetto al numero totale di casi nel mondo Asia
percentage_asia = (df_asia['new_cases'].sum() / total_world_cases) * 100
print("Percentuale rispetto al numero totale di casi nel mondo continente Asia:")
percentage_asia
```

```python
# Converisone colonna 'date' in formato datetime
covid_data['date'] = pd.to_datetime(covid_data['date'])

# Filtro dati per l'Italia nel 2022
italy_data_2022 = covid_data[(covid_data['location'] == 'Italy') & (covid_data['date'].dt.year == 2022)]

# Visualizzazione prime 5 righe per controllo del dataframe filtrato
italy_data_2022.head()
```

```python
# Generazione grafico con metodo plot
italy_data_2022.plot(x='date', y='total_cases', figsize=(10, 6), title ='Evoluzione dei casi totali in Italia nel 2022', xlabel='Data', ylabel='Casi Totali')
```

```python
# Generazione grafico con metodo plot
italy_data_2022.plot(x='date', y='new_cases', figsize=(10, 6), title ='Numero di nuovi casi in Italia nel 2022', xlabel='Data', ylabel='Nuovi Casi Giornalieri')
```

```python
# Calcolo della somma cumulativa dei nuovi casi giornalieri
italy_data_2022.loc[:, 'cumulative_new_cases'] = italy_data_2022['new_cases'].cumsum()
```

```python
# Generazione del grafico con Matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
italy_data_2022.plot(x='date', y='cumulative_new_cases', marker='o', linestyle='-', figsize=(10, 6))
plt.title('Numero cumulativo dei nuovi casi in Italia nel 2022')
plt.xlabel('Data')
plt.ylabel('Numero cumulativo dei nuovi casi')
plt.grid(True)
plt.xticks(rotation=45)

# Commento
plt.text(0.5, -0.3, "Il grafico della somma cumulativa dei nuovi casi del 2022 in Italia riflette in gran parte l'andamento dei casi totali nel periodo. Questo mi fa capire che i numeri in total_cases erano appunto progressivi / cumulativi", transform=plt.gca().transAxes, ha='center', fontsize=10,wrap=True) 
plt.show()
```

```python
import seaborn as sns

selected_countries = ['Italy', 'Germany', 'France']
start_date = '2022-05-01'
end_date = '2023-04-30'

# Filtro dati per le nazioni specificate nel periodo specificato
filtered_data = covid_data[((covid_data[ 'location' ] == 'Italy') |
                           (covid_data[ 'location' ] == 'Germany') |
                           (covid_data[ 'location' ] == 'France')) &
                           ((covid_data[ 'date' ] >= start_date) &
                           (covid_data[ 'date' ] <= end_date))]

# Controllo del dataframe filtrato
filtered_data
```

```python
# Controllo esito somme
icu_patients_per_country = filtered_data.groupby( 'location' )['icu_patients'].sum()
print("Pazienti in terapia intensiva per nazione:")
icu_patients_per_country
```

```python
# Generazione del boxplot con Seaborn
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='location',y='icu_patients_per_million', data=filtered_data)
plt.title('Differenza nel numero di pazienti in terapia intensiva tra Italia, Germania e Francia (Maggio 2022 - Aprile 2023)')
plt.xlabel('Nazione')
plt.ylabel('Numero di pazienti in terapia intensiva')
plt.show()
```
